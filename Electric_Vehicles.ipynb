```python
# Compare Range vs Battery
rb=px.scatter(df,x='battery_capacity_kWh',y='range_km',size='range_km',size_max=30,color='range_km',hover_data=['brand', 'model'])
rb.update_layout(title=' how battery capacity affects driving range.')
rb.show()
```

```python
top_brands = df['brand'].value_counts().nlargest(10)
print(top_brands)
```

```python
##### You get a horizontal bar chart showing the top 10 brands with the most electric vehicle models in dataset, with darker shades indicating higher counts.
```

```python
# Count the number of models per brand
top_brands = df['brand'].value_counts().nlargest(10).reset_index()     # Count the number of models per brand
top_brands.columns = ['brand', 'count']                                # Rename columns for clarity

# Plot
fig1 = px.bar(
    top_brands,
    x='count',
    y='brand',
    orientation='h',
    title='Top 10 Brands by Number of Models',
    labels={'count': 'Number of Models', 'brand': 'Brand'},
    color='count',
    color_continuous_scale='Blues'
)
fig1.show()
```

```python
#Analyze Acceleration by Car Segment
box=px.box(df, x='segment', y='acceleration_0_100_s')
box.show()
```

```python
#Car Range Distribution
#Goal: See how most EVs are distributed in terms of range.

# Step 2: Create histogram with hover data
fig =px.histogram(
    df,
    x='range_km',
    nbins=20,
    title='Distribution of EV Range with Brand and Model Info',
    color_discrete_sequence=['skyblue'],

)
fig.show()
```

```python
#3D View of EV Performance
#Goal: Explore the relationship between acceleration, top speed, and range
dg=px.scatter_3d(df,x='top_speed_kmh',y='acceleration_0_100_s',z='range_km',color='segment')
dg.update_layout(title='3D View of EV Performance')
dg.show()
```

```python
#Range by Drivetrain
#Goal: See how range varies by drivetrain (FWD, RWD, AWD).
dt=px.box(df,x='drivetrain',y='range_km',)
dt.show()

```

```python
#Battery Efficiency Comparison by Car Body Type
#Goal: Check which body types are most energy efficient.
v=px.violin(df,x='car_body_type', y='efficiency_wh_per_km',box=True)
v.show()


```
